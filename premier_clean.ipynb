```python
import pandas as pd
df = pd.read_csv('matches.csv')
display(df.head(10))

```

```python
#Aparentemente, las columnas de notes y match report no están aportando información relevante.
# Match report solo tiene match report en texto, y notes tiene valores nulos. Verificamos
display(f"Hay {df['notes'].isna().sum()} valores nulos en la columna notes")
display(f"Hay {df['match report'].isna().sum()} valores nulos en la columna match report")
display(f"Valores únicos en la lista: {df['match report'].unique().tolist()}")
df.drop('notes', axis=1, inplace=True)
df.drop('match report', axis=1, inplace=True)
df.head(5)
```

```python
df.info()
# Podemos observar que hay valores nulos en la columna attendance, reemplazamos por 0 esos valores
df['attendance'] = df['attendance'].fillna(0)
```

```python
# Verificamos que la lista de equipos no tenga equipos repetidos que puedan ser el mismo equipo
display(pd.DataFrame(df['team'].unique().tolist(), columns = ['teams']).sort_values('teams'))
display(pd.DataFrame(df['opponent'].unique().tolist(), columns = ['opponents']).sort_values('opponents'))
teams = set(df['team'].unique().tolist())
opponents = set(df['opponent'].unique().tolist())
display(teams)
display(opponents)
display(f"La diferencia de equipos es: {teams.difference(opponents)}")
display(f"La diferencia de equipos es: {opponents.difference(teams)}")
#Notaremos mejor las diferencias de los hacemos las diferencias A-B y B-A de los conjuntos teams y opponents
#Podré ver la "version" distinta de los equipos

#La columna unnamed 0, no aporta información relevante, la eliminamos
#df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
# La columna comp, no aporta información relevante, la eliminamos, la competición es la premier league, ya sabemos que son partidos de premier league
df.drop('comp', axis=1, inplace=True)
variables = df.columns
display(variables)
```

```python
df['attendance'] = df['attendance'].fillna(0)
display(df['attendance'].unique())
```

```python
display(df.info())
df.head(10)

```

```python
#Observemos los días únicos
display(df['day'].unique())
display(df['venue'].unique())
display(df['formation'].unique())

#Ya tenemos que hay que corregir las formaciones, y también corregir los teams como se mencionó anteriormente

df['opponent'] = df['opponent'].replace('Tottenham', 'Tottenham Hotspur')
df['opponent'] = df['opponent'].replace("Nott'ham Forest", 'Nottingham Forest')
df['opponent'] = df['opponent'].replace('West Brom', 'West Bromwich Albion')
df['opponent'] = df['opponent'].replace('Brighton', 'Brighton and Hove Albion')
df['opponent'] = df['opponent'].replace('West Ham', 'West Ham United')
df['opponent'] = df['opponent'].replace('Sheffield Utd', 'Sheffield United')
df['opponent'] = df['opponent'].replace('West Ham', 'West Ham United')
df['opponent'] = df['opponent'].replace('Newcastle Utd', 'Newcastle United')
df['opponent'] = df['opponent'].replace('Manchester Utd', 'Manchester United')
df['opponent'] = df['opponent'].replace('Wolves', 'Wolverhampton Wanderers')

teams = set(df['team'].unique().tolist())
opponents = set(df['opponent'].unique().tolist())

df['formation'] = df['formation'].str.replace(r'[^\x00-\x7F]+', '', regex=True)
df['formation'] = df['formation'].str.rstrip('-0')

display(teams.difference(opponents)) # Verificamos que en efecto, ya no hay diferencia entre teams y opponents
```

```python
df.head(10)
```

```python
display(df['captain'].unique())
display(df['referee'].unique())
display(df['result'].unique())
```

```python
# Round podemos dejarla entera

df['round'] = df['round'].astype(str).str.lstrip('Matchweek ')
df['round'] = df['round'].astype(int)
df.head(10)

```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['year'] = df['date'].dt.year
df.head(10)

#Hay algo sospechoso en los datos, la fecha de los partidos no corresponde con la temporada de la premier league
#Lo dejamos pasar

#Deben presentarse dos horas por la diferencia horaria de la premier league, separamos las horas en dos columnas

#display(df['time_zone1'].unique())
#df['time_zone2'] = df['time'].str[1]

display(df['time'].unique()) #Reviso si hay algo raro en los datos

df['time_zone1'] = df['time'].str.split('(').str[0].str.strip()
df['time_zone2'] = df['time'].str.split('(').str[1].str.strip(')')

#Convierto tipo de dato de las columnas
df['time_zone1'] = pd.to_datetime(df['time_zone1'], format='%H:%M').dt.strftime('%H:%M')
df['time_zone2'] = pd.to_datetime(df['time_zone2'], format='%H:%M').dt.strftime('%H:%M')


display(df[['time_zone1', 'time_zone2']].head(10))



```

```python
df.drop('time', axis=1, inplace=True)
```

```python
df.info()
```

```python
#Había ignorado dos missing values de dist, los reemplazo por 0
df['dist'] = df['dist'].fillna(0)
df.info()
```

```python
#Hemos confirmado que no hay valores nulos, la primera columna Unnamed está haciendo ruido, no
#sabemos para que sirve, ni que representa, la eliminamos

df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#Guardamos el archivo limpio
df.to_csv('matches_clean.csv', index=False)
```
